#### 사이킷런의 샘플 데이터 살펴보기 - 샘플 코드

```python
import pandas as pd
from sklearn.datasets import load_iris

data = load_iris()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.DataFrame(data.target, columns=['Species'])

df = pd.concat([X, y], axis=1)
df.head()
```

#### 지도 학습(분류)의 예 - 샘플 코드 1

```python
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
```

#### 지도 학습(분류)의 예 - 샘플 코드 2

```python
X = data.data
y = data.target
```

#### 지도 학습(분류)의 예 - 샘플 코드 3

```python
X = X[:, :10]
```

#### 구현 방법 - 샘플 코드 1

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs')
```

#### 구현 방법 - 샘플 코드 2

```python
model.fit(X, y)
```

#### 구현 방법 - 샘플 코드 2

```python
y_pred = model.predict(X)
```

#### 구현 방법 - 샘플 코드 3

```python
from sklearn.metrics import accuracy_score

accuracy_score(y, y_pred)
```

#### 비지도 학습의 예 - 샘플 코드 1

```python
from sklearn.datasets import load_wine

data = load_wine()
```

#### 비지도 학습의 예 - 샘플 코드 2

```python
X = data.data[:, [0, 9]]
```

#### 비지도 학습의 예 - 샘플 코드 3

```python
from sklearn.cluster import KMeans

n_cluster = 3
model = KMeans(n_clusters=n_cluster)
```

#### 비지도 학습의 예 - 샘플 코드 4

```python
pred = model.fit_predict(X)
```

#### 비지도 학습의 예 - 샘플 코드 5

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.scatter(X[pred==0, 0], X[pred==0, 1], color='red', marker='s', label='Label1')
ax.scatter(X[pred==1, 0], X[pred==1, 1], color='blue', marker='s', label='Label2')
ax.scatter(X[pred==2, 0], X[pred==2, 1], color='green', marker='s', label='Label3')
ax.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=200, color='yellow', marker="*", label="center")

ax.legend()
plt.show()
```

#### 그래프를 인라인에 출력 - 샘플 코드 1

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

#### 그래프를 인라인에 출력 - 샘플 코드 2

```python
x1 = np.linspace(-5, 5, 101)
y1 = np.sin(x1)
```

#### 그래프를 인라인에 출력 - 샘플 코드 3

```python
plt.plot(x1, y1)
```

#### 그래프를 인라인에 출력 - 샘플 코드 4

```python
fig, ax = plt.subplots()

ax.set_title('Sin')
ax.set_xlabel('rad')
ax.plot(x1, y1)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
plt.show()
```
